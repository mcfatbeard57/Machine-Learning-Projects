```python
a = [-13, 3, 25, 20, -3, 0, 2, 17, 5, 22, -15, -4, -7]
# maximum sum of contiguous subarray
```

```python
# Pick any syb array of any length
# sub array will be continous i.e. all elements are in order and not skipped over
# find its sum
# find the array with max sum of any size
```

```python
# Condition it has to be continous so thats the way to start
```

```python
for i in range(len(a)):
    count = 0
    while count < len(a):
        count+=1
        #if i < len(a):
        sum = 0
        for j in a[i+1 : count]:
            sum = sum + j
        print(a[i] + sum)
    print('+'*20)
```
